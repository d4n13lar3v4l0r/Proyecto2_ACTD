```python
data.rename(columns={'X1_LIMIT_BAL':'Saldo_Limite', 'X2_SEX':'Sexo','X3_EDUCATION':'Educacion','X4_MARRIAGE':'Estado_Civil'},inplace=True)
```

```python
#SEXO (1= Hombre, 2 = Mujer)

data['Sexo'].value_counts()
```

```python
#Nivel de Educación (1 = Posgrado, 2 = Pregrado, 3 = Bachillerato, 4 = Otros)

data['Educacion'].value_counts()
```

```python
#Hay valores como 5, 6 y 0 que son desconocidos. Se le agregarán a la categoría 4 "Otros"

fil = (data['Educacion'] == 5) | (data['Educacion'] == 6) | (data['Educacion'] == 0)
data.loc[fil, 'Educacion'] = 4
data['Educacion'].value_counts()
```

```python
#Estado Civil (1 = Casadx, 2 = Solterx, 3 = Otro)

data['Estado_Civil'].value_counts()

```

```python
#Hay valores 0 que son desconocidos. Se le agregarán a la categoría 3 "Otro"

fil = data['Estado_Civil'] == 0
data.loc[fil, 'Estado_Civil'] = 3
data['Estado_Civil'].value_counts()
```

Debido a que el dataset está desbalanceado, se realiza una corrección por SMOTE (Synthetic Minority Oversampling Technique)

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(data.iloc[:,0:-1], data['EsDefault'])

print('Original dataset shape', len(data))
print('Resampled dataset shape', len(y_smote))
```

```python
x_smote
```

```python
columnas = list(data.columns)
```

```python
columnas.pop()
```

```python
df_balanceado = pd.DataFrame(x_smote, columns=columnas)
```

```python
df_balanceado['EsDefault'] = y_smote
```

```python
sns.countplot(x='EsDefault', data = df_balanceado, palette="Set3")
```
